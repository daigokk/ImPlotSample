```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1,2, figsize=(6,4))
dat_raw = np.loadtxt('01_raw.csv', delimiter=',')
dat_fft = np.loadtxt('02_fft.csv', delimiter=',')
ax[0].plot(dat_raw[:,0]*1e3, dat_raw[:,1])
ax[0].set_title('Raw Signal')
ax[0].set_xlabel('Time [ms]')
ax[0].set_ylabel('Volatege (V)')
ax[1].plot(dat_fft[:,0]*1e-6, dat_fft[:,1])
ax[1].set_xlim(0, 1/(dat_raw[1,0]-dat_raw[0,0])/2*1e-6)
ax[1].set_title('FFT of Signal')
ax[1].set_xlabel('Frequency [MHz]')
ax[1].set_ylabel('Magnitude (V)')
fig.tight_layout()
fig.savefig('signal_analysis.svg')
```
